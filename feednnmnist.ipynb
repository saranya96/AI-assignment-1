```python
# %load cln2.py
import tensorflow as tf
 
# importing data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)

 
#setting parameters
learn_rate      = 0.003
train_epochs = 40
batch_size      = 100
 
 
#network parameters
nh1= 1000 # nr of neurons in 1st hidden layer
nh2 = 1000 # nr of neurons in 2nd hidden layer
n_input    = 784 # MNIST data input size (one input image has dimension 28x28 pixels, thus 784 input pixels)
numclass =  10 # MNIST total classes (0-9 digits)
 
 
# 4
x = tf.placeholder("float", [None, n_input])
y = tf.placeholder("float", [None, numclass])
 
 

# input-layer --> hidden layer1 --> hidden layer2 --> output layer
def mlayerpercep(x, weights, biases): #multilayered
 
    # hidden layer 1
    hl1= tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    hl1= tf.nn.relu(hl1)  #reLU activation
     
    # hidden layer 2 
    hl2 = tf.add(tf.matmul(hl1, weights['h2']), biases['b2'])
    hl2 = tf.nn.relu(hl2)
     
    # output layer
    outputlayer = tf.matmul(hl2, weights['out']) + biases['out']
     
   
    return outputlayer
 
     
# weights & biases of the layers 
weights = {
    'h1': tf.Variable(tf.random_normal([n_input, nh1])),
    'h2': tf.Variable(tf.random_normal([nh1, nh2])),
    'out': tf.Variable(tf.random_normal([nh2, numclass]))
}
biases = {
    'b1': tf.Variable(tf.random_normal([nh1])),
    'b2': tf.Variable(tf.random_normal([nh2])),
    'out': tf.Variable(tf.random_normal([numclass]))
}
 
 

my_mlp = mlayerpercep(x, weights, biases)
 
 
# loss function
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=my_mlp, logits=y))
 
 
# optimizer
optimizer = tf.train.AdamOptimizer(learning_rate=learn_rate).minimize(cost)
 
 

init = tf.global_variables_initializer()
sess = tf.InteractiveSession()
sess.run(init)
 
 
#training
for epoch in range(train_epochs):
 
   
    epoch_cost = 0.0
 
   
    nr_batches_to_process = int(mnist.train.num_examples/batch_size)
     
   
    for i in range(nr_batches_to_process):
     
        
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        
       
        _, c = sess.run([optimizer, cost], feed_dict={x: batch_x,
                                                      y: batch_y})
                                                       
        # compute total cost 
        epoch_cost += c
     
    # cost for each epoch
    print("Epoch:", '%03d' % epoch, ", epoch cost=", "{:.3f}".format(epoch,epoch_cost))
     
print("Done")
 
# 12. test the model
correct_prediction = tf.equal(tf.argmax(my_mlp, 1), tf.argmax(y, 1))
 
# 13. calculate accuracy of the learned model
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float")) 
print("Accuracy:", accuracy.eval({x: mnist.test.images, y: mnist.test.labels}))
```
